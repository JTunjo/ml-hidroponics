```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyarrow
```

```python
df = pd.read_csv("cafe_nov_2025.csv", low_memory=False)

cols_string = [
    "C112 DESCRIPCION",
    "C115 COD. PAS ORIGEN",
    "C68 CD. TIPO CARGA"
]

for col in cols_string:
    df[col] = df[col].astype(str)

# 3️⃣ Convertir FOB a numérico (forzar errores a NaN)
df["C72 VALOR TOTAL FOB U$"] = pd.to_numeric(
    df["C72 VALOR TOTAL FOB U$"], errors="coerce"
)
```

```python
# 4️⃣ Aplicar filtros
df_filtrado = df[
    (df["C112 DESCRIPCION"].str.contains("tostado", case=False, na=False) &
    df["C115 COD. PAS ORIGEN"].str.contains("colombia", case=False, na=False) &
    df["C68 CD. TIPO CARGA"].str.contains("contenerizada", case=False, na=False) &
    df["C53 CD. TIPO DATOS"].str.contains("definitivo", case=False, na=False)) &
    (df["C72 VALOR TOTAL FOB U$"].notna()) &
    (df["C49 RGIMEN ADUANERO"] == 42)
]

# 5️⃣ Ver resultado
print("Filas originales:", len(df))
print("Filas filtradas:", len(df_filtrado))
```

```python
df_caja = df_filtrado[
    (df_filtrado["C106 COD. CLASE EMBALAJE"].str.contains("caja", case=False, na=False))
    ]
```
